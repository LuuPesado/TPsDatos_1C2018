```python
import datetime as datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
postulantes_educacion = pd.read_csv('datos_navent_fiuba/fiuba_1_postulantes_educacion.csv')
postulantes_genero_edad = pd.read_csv('datos_navent_fiuba/fiuba_2_postulantes_genero_y_edad.csv')
vistas = pd.read_csv('datos_navent_fiuba/fiuba_3_vistas.csv')
postulaciones = pd.read_csv('datos_navent_fiuba/fiuba_4_postulaciones.csv')
avisos_online = pd.read_csv('datos_navent_fiuba/fiuba_5_avisos_online.csv')
avisos_detalles = pd.read_csv('datos_navent_fiuba/fiuba_6_avisos_detalle.csv')
```

```python
postulantes = postulantes_educacion.join(postulantes_genero_edad.set_index('idpostulante'), on='idpostulante',how='left')
```

```python
##unimos los avisos para agregar la columna online en avisos_detalles
avisos = avisos_detalles.join(avisos_online.set_index('idaviso'),on='idaviso',how='left')
```

```python
avisos
```

```python
##Los que no son avisos online la fila en la columna online tiene valor False
avisos["online"].fillna(False,inplace=True)
avisos
```

```python
## Los 5 avisos mas vistos en todo el dataframe de visitas
vistas.loc[:,["idAviso","idpostulante"]].groupby("idAviso").agg('size').sort_values(ascending=False).head()
```

```python
## Reemplazamos los IdAviso por el area de los postulantes
vistas_all = vistas.join(avisos_detalles.set_index("idaviso"), on="idAviso", how='inner')
vistas_all.loc[:,["nombre_area", "idpostulante"]].groupby("nombre_area").agg(['size'])["idpostulante"].sort_values(by="size",ascending=False)
```

```python
##cantidad de postulaciones segun el area
vistas_area = vistas_all.loc[:,["nombre_area", "idpostulante"]].groupby("nombre_area").agg(['size'])["idpostulante"].sort_values(by='size',ascending=False)
vistas_area.plot(kind='bar',figsize=(16,8))
plt.title("Cantidad de visitas vs Area")
plt.xlabel("Area")
plt.ylabel("Candidad de visitas")
```

```python
##Las 20 areas mas vistas por postulantes en todo el dataframe
vistas_area = vistas_area.head(20)
vistas_area.rename(columns={'nombre_area':'Area'}, inplace=True)
vistas_area.plot(kind='bar',title="Cantidad de visitas VS Area")
plt.xlabel("Area")
plt.ylabel("Cantidad de visitas")
```

```python
##Filtramos por tiempo. Primeros 5 avisos mas vistos en el último mes(Marzo)
vistas["timestamp"] = pd.to_datetime(vistas["timestamp"], errors='coerce')
vistas_ultimo_mes = vistas.loc[vistas["timestamp"].dt.month==3,:]
vistas_ultimo_mes.loc[:,["idAviso","idpostulante"]].groupby("idAviso").agg('size').sort_values(ascending=False).head()
```

```python
##tipo de aviso que mas visitado en el ultimo mes(Marzo)
avisos.loc[avisos["idaviso"]==1112369306,["nombre_area","denominacion_empresa"]]
```

```python
##Segundo aviso mas visitado en el último mes(Marzo)
avisos.loc[avisos["idaviso"]==1112368053,["nombre_area","denominacion_empresa"]]
```

```python
##Grafico de los avisos mas visitados en el último mes (Marzo)
avisos_ultimo_mes = vistas_ultimo_mes.join(avisos_detalles.set_index("idaviso"), on="idAviso", how='inner')
avisos_ultimo_mes.loc[:,["nombre_area","idpostulante"]].groupby("nombre_area").agg([np.size])["idpostulante"]\
            .sort_values(by="size",ascending=False).head(20).plot(kind="bar", title="Cantidad de postulaciones VS Area")
plt.xlabel("Area")
plt.ylabel("Cantidad de visitas")
```

```python
##Primeros 5 avisos online mas vistos en el ultimo mes 
vistas_ultimo_mes_online = vistas.loc[vistas["timestamp"].dt.month==3,:]
vistas_ultimo_mes_online = vistas_ultimo_mes_online.join(avisos_online.set_index("idaviso"),on='idAviso', how='inner')
vistas_ultimo_mes_online = vistas_ultimo_mes_online.loc[vistas_ultimo_mes_online["online"]==True,:]
vistas_ultimo_mes_online = vistas_ultimo_mes_online.join(avisos_detalles.set_index("idaviso"),on="idAviso", how="inner") 
vistas_ultimo_mes_online.loc[:,["idAviso","idpostulante","nombre_area"]].groupby("idAviso")\
    ["idpostulante"].agg([np.size]).sort_values(by="size",ascending=False).head()
```

```python
##Grafico de Barras de los 20 avisos mas vistos segun el Area
vistas_ultimo_mes_online[["nombre_area","idpostulante"]].groupby("nombre_area").agg('size')\
            .plot(kind="bar", title="Cantidad de visitas online vs area del aviso", figsize=(16,8))
plt.ylabel("Candidad de visitas")
plt.xlabel("Area del aviso")
```

```python
##Filtramos los anuncios por el mas visitado en el mes
id_post = vistas.loc[vistas["idAviso"]==1112369306,["idAviso","idpostulante"]]
id_post
```

```python
##Unimos por el id del postulantes con el dataframe de postulaciones y graficamos la cantidad de personas que visitaron
##el anuncio mas visto según su educación
id_post=id_post.join(postulantes.set_index("idpostulante"), on="idpostulante", how="inner")
id_post["educacion"].value_counts().plot(kind="bar", title="Cantidad de personas según educación", rot=0, figsize=(16,8))
plt.xlabel("Tipo de educación")
plt.ylabel("Cantidad de postulantes")
```

```python
##vemos cuántos terminaron el secundario y cuantos no
id_post.loc[id_post["educacion"]=="Secundario",["educacion","estado"]].groupby("estado").agg([np.size])
```

```python
##Renombramos la columna nombre por educacion
postulantes = postulantes.rename(index=str,columns={"Educacion":"educacion"})
postulantes
```

```python
##Que educacion posee el que mas avisos visita
post_edu = postulantes.join(vistas.set_index("idpostulante"), on="idpostulante", how='inner')
post_edu.loc[:,["educacion","idAviso"]].groupby("educacion").agg('size').sort_values(ascending=False)\
        .plot(title="Cantidad de visitas según la educación del postulante", kind="bar", figsize=(16,8), rot=0)
plt.xlabel("Educacion")
plt.ylabel("Cantidad de Visitas")
```
