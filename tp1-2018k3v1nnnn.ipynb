## Cantidad de visitas segun el dia de la semana

```python
def separar_fecha(fecha):
    nueva_fecha=fecha.split("T")
    return nueva_fecha[0]
def separar_horario(fecha):
    nueva_fecha=fecha.split("T")
    dato=nueva_fecha[1]
    nuevo_horario=dato.split("-")
    return nuevo_horario[0]
```

```python
postulantes["nueva_fecha"]=postulantes["timestamp"].map(separar_fecha)
postulantes["nuevo_horario"]=postulantes["timestamp"].map(separar_horario)
```

```python
postulantes
```

```python
def dia_de_semana(fecha):
    return calendar.day_name[fecha.weekday()]
    
```

```python
postulantes["nueva_fecha"]=pd.to_datetime(postulantes["nueva_fecha"])
postulantes["nuevo_horario"]=pd.to_datetime(postulantes["nuevo_horario"])
```

```python
postulantes["dia"]=postulantes["nueva_fecha"].map(dia_de_semana)
postulantes["hora"]=postulantes["nuevo_horario"].dt.hour
```

```python
postulantes
```

```python
postulantes["dia"].value_counts()
```

```python
postulantes["dia"].value_counts().plot(kind="bar",rot=0,figsize=(10,9),color="aqua",fontsize=13);
plt.title('Cantidad de visitas segun el dia', fontsize=22);
plt.xlabel('Dia de la semana', fontsize=18);
plt.ylabel('Cantidad de visitas', fontsize=20);
```
