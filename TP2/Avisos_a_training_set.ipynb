```python
del avisos['Unnamed: 0']
```

```python
avisos['nombre_zona'].value_counts()
```

```python
def zona_to_num(zona):
    if zona == 'Gran Buenos Aires':
        return 0
    if zona == 'Capital Federal':
        return 1
    else:
        return 2
avisos['nombre_zona'] = avisos['nombre_zona'].map(lambda z: zona_to_num(z))
```

```python
avisos['tipo_de_trabajo'].value_counts()
```

```python
keys = avisos['tipo_de_trabajo'].value_counts().keys()
dict_tipo_to_num = {}
i = 0
for k in keys:
    dict_tipo_to_num[k] = i
    i = i + 1
dict_tipo_to_num
```

```python
avisos['tipo_de_trabajo'] = avisos['tipo_de_trabajo'].map(lambda t: dict_tipo_to_num[t])
```

```python
avisos['nivel_laboral'].value_counts()
```

```python
avisos['nivel_laboral'] = avisos['nivel_laboral'].fillna('0')
nivel_to_num = {'Otro': 0,
                'Junior': 1,
                'Senior / Semi-Senior':2,
                'Jefe / Supervisor / Responsable':3,
                'Gerencia / Alta Gerencia / Dirección':4,
                '0' : 0
                }
avisos['nivel_laboral'] = avisos['nivel_laboral'].map(lambda n: nivel_to_num[n])
```

```python
avisos['nombre_area'].value_counts().keys()
```

```python
keys = avisos['nombre_area'].value_counts().keys()
dict_nombre_area = {}
i = 0
for k in keys:
    dict_nombre_area[k] = i
    i = i + 1
dict_nombre_area
```

```python
avisos['nombre_area'] = avisos['nombre_area'].map(lambda n: dict_nombre_area[n])
```
